new idea:

song shuffler

person inputs up to 10 songs, stored in a list

procedure: inputs list of songs, then uses a for loop to pick random song, and adds it to list of shuffled songs.   
but same song cannot be used twice in a row

```python
import random
import time

songs = []
add_new_song = ''


while not add_new_song.lower() == 'no':
    add_new_song = input("What song do you want to add to the playlist? If you don't want to add any more songs, just enter 'no.' ")

    songs.append(add_new_song)

len_shuffled = int(input("How long would you like your shuffled list to be? "))


def shuffle_songs(songs_list):
    shuffled_songs = []

    print("Shuffling songs...")
    time.sleep(0.5)

    while len(shuffled_songs) < len_shuffled:
        song_to_add = random.choice(songs_list)

        if len(shuffled_songs) == 0:
            shuffled_songs.append(song_to_add)

        elif song_to_add == shuffled_songs[-1]:
            pass

        else:
            shuffled_songs.append(song_to_add)

    return shuffled_songs

print("Shuffled songs: " + str(shuffle_songs(songs)))
```
