```python
import random
import time

songs = []
add_new_song = ''

while not add_new_song.lower().strip() == 'no':
    add_new_song = input("What song do you want to add to the playlist? If you don't want to add any more songs, just enter 'no.' ")

    if add_new_song.lower().strip() != 'no':
      songs.append(add_new_song)
    else:
      continue

len_shuffled = int(input("How long would you like your shuffled list to be? "))
print()

def shuffle_songs(songs, len_shuffled):
    shuffled_songs = []

    print("Shuffling songs...")
    time.sleep(0.5)
    if len(songs) > 0:
      while len(shuffled_songs) < len_shuffled:
          song_to_add = random.choice(songs)

          if len(shuffled_songs) == 0:
              shuffled_songs.append(song_to_add)

          elif song_to_add == shuffled_songs[-1] and len(songs) != 1:
              pass

          else:
              shuffled_songs.append(song_to_add)

      return shuffled_songs
    else:
      print()
      print("I can't shuffle songs from an empty list!")

result_playlist = shuffle_songs(songs, len_shuffled)

if result_playlist is not None:
    print()
    print("Shuffled songs: " + str(result_playlist))
```
